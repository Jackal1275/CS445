```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline

redData = pd.read_csv("winequality-red.csv")
whiteData = pd.read_csv("winequality-white.csv")

#  P. Cortez, A. Cerdeira, F. Almeida, T. Matos and J. Reis. 
#  Modeling wine preferences by data mining from physicochemical properties.
#  In Decision Support Systems, Elsevier, 47(4):547-553. ISSN: 0167-9236.
print("Red Wine Data")
display(redData.head())
print("White Wine Data")
display(whiteData.head())
```

```python
red_quality_by_avg_acidity = redData.groupby(['quality'])['fixed acidity'].mean()
print(red_quality_by_avg_acidity)
```

```python
plt.figure(figsize=(8,6))
red_quality_by_avg_acidity.plot(kind='bar')
plt.xlabel('Quality')
plt.ylabel('Average Acidity')
plt.title('Average Fixed Acidity by Quality')
#plt.legend(title='Smoker')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
